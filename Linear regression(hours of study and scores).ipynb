```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/student_scores.csv')
```

```python
plt.scatter(data.Hours,data.Scores,color="black")
```

```python
def gradient_descent(m_current,b_current,points,L):
  m_gradient=0
  b_gradient=0
  n=len(points)
  for i in range(n):
    x=points.iloc[i].Hours #to get the value according to the index -iloc is index location
    y=points.iloc[i].Scores
    m_gradient+=-(2/n)*x*(y-(m_current*x+b_current))
    b_gradient+=-(2/n)*(y-(m_current*x+b_current))
  m=m_current-m_gradient * L
  b=b_current-b_gradient * L
  return m,b
  
```

```python
m=1
b=0
L=0.0001
epochs=1000
for i in range(epochs):
  if i%50==0:
    print(f"Epoch: {i}")
  m,b=gradient_descent(m,b,data,L)
print(m,b)
```

```python
plt.scatter(data.Hours,data.Scores,color="black",marker="o")
plt.plot(list(range(2,10)),[m*x+b for x in range(2,10)],color="red")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("LINEAR REGRESSION OF SCORES BASED ON STUDY HOURS")
```
